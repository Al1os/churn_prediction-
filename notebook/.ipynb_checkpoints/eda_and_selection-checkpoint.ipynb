```python
from app.scraper import load_data
from app.features import create_features

df = load_data()

df = create_features(df)

print(df.shape)

df.head()
```

Method 1: Filter

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(
    threshold=0.01
)
X = df.drop(columns=["churned"])
y = df["churned"]

print("Original Features:")
print(X.columns.tolist())

X_var = selector.fit_transform(X)
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(
    X.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.show()
```

```python
corr_matrix = X.corr().abs()

upper_triangle = corr_matrix.where(
    np.triu(
        np.ones(corr_matrix.shape),
        k=1
    ).astype(bool)
)

high_corr_features = [
    column
    for column in upper_triangle.columns
    if any(upper_triangle[column] > 0.90)
]

print("\nFeatures removed because correlation > 0.90:")
print(high_corr_features)

X_corr_filtered = X.drop(
    columns=high_corr_features
)

print("\nFeatures after correlation filtering:")
print(X_corr_filtered.columns.tolist())
```

Method 2: RFE

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000
)

rfe = RFE(
    estimator=model,
    n_features_to_select=5
)

rfe.fit(X, y)

print(
    X.columns[rfe.support_]
)
```

Method 3: Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=5
)

dt.fit(X,y)

dt.feature_importances_
```

Method 4: Random forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(X,y)

rf.feature_importances_
```
